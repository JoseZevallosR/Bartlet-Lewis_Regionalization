```python
import pandas as pd
import geopandas as gpd
import folium
import os
import sqlite3
import matplotlib.pylab as plt
from scipy import stats
import numpy as np
import haversine as hs
from haversine import Unit
%config Completer.use_jedi = False #for auto complete code
```

```python
maestro=pd.read_excel('../data/Maestro/maestro.xlsx')
```

```python
area_estudio=pd.read_csv('../data/Area_estudio/gauges.csv',sep=';')
```

```python
area_estudio.head()
```

```python
maestro.head()
```

```python
maestro[['NOMBRE_ESTACION','LONGITUD','LATITUD']].head()
```

```python
name_idx=[]
min_distances=[]
for idx, rowg in area_estudio.iterrows():
    loc1=rowg.x,rowg.y
    #loc1=area_estudio.x[0],area_estudio.y[0]
    distance=[]
    for index,row in maestro.iterrows():
        loc2=(row.LONGITUD,row.LATITUD)
        distance.append(hs.haversine(loc1,loc2,unit=Unit.METERS))
      
    name_idx.append(distance.index(np.min(distance)))
    min_distances.append(np.min(distance))

    
```

```python
area_estudio['Nombre']=maestro.loc[name_idx,'NOMBRE_ESTACION'].to_list()
area_estudio['ALTITUD']=maestro.loc[name_idx,'ALTITUD'].to_list()
area_estudio['DEPARTAMENTO']=maestro.loc[name_idx,'DEPARTAMENTO'].to_list()
area_estudio['PROVINCIA']=maestro.loc[name_idx,'PROVINCIA'].to_list()
area_estudio['TIPO']=maestro.loc[name_idx,'TIPO'].to_list()
area_estudio['CATEGORIA']=maestro.loc[name_idx,'CATEGORIA'].to_list()
area_estudio['CODIGO']=maestro.loc[name_idx,'CODIGO'].to_list()
```

```python
area_estudio.to_csv('estaciones_estudio.csv',index=False)
```
